```python
import pandas as pd


df1 = pd.DataFrame({
    'Name': ['A', 'A', 'C', 'D'],
    'Marks': [10, 20, 30, 40]
})


df2 = pd.DataFrame({
    'Name': ['A', 'A', 'A', 'E', 'F'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on=['Name'],how='inner')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Name',how='left')
```

```python
pd.merge(df1,df2,on='Name',how='right')
```

```python
pd.merge(df2,df1,on='Name',how='left',indicator=True)
```
